```python
from sklearn.datasets import load_iris # load iris data set from sklearn
from sklearn.decomposition import PCA # PCS is a technique for reducing the dimensionality of a dataset while retaining most of the information in the data.
import matplotlib.pyplot as plt 

data = load_iris () # load iris data into data
X= data.data # load only data values in X
y= data.target # load the target values in y

pca = PCA (n_components=2) # first we should define PCA componenet
x_pca = pca.fit_transform(X) # "function.fit_transform (data)" this command commonly use for PCA or Standardization of dataset.
print (x_pca[:, 0])  # [:, 0] this means all data in column 1
print (x_pca[:, 1]) # [:, 1] this means all data in column 2
print (y) 


plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
from sklearn.datasets import load_iris # load iris data set from sklearn
from sklearn.decomposition import PCA # PCS is a technique for reducing the dimensionality of a dataset while retaining most of the information in the data.
import matplotlib.pyplot as plt 

data = load_iris () # load iris data into data
X= data.data # load only data values in X
y= data.target # load the target values in y

pca = PCA (n_components=3) # first we should define PCA componenet
x_pca = pca.fit_transform(X) # "function.fit_transform (data)" this command commonly use for PCA or Standardization of dataset.

print (x_pca[:, 2])

plt.scatter( x_pca[:, 0], x_pca[:, 2],  c=y)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
